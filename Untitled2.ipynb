```python
import numpy as np
```

```python
Energy = np.loadtxt('Energy.dat')
```

```python
Energy = Energy.transpose() + Energy
```

```python
for i in range(Energy.shape[0]):
    Energy[i, i] *= .5
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(Energy)
```

```python
Entropy = np.loadtxt('Entropy.dat')
Entropy = Entropy.transpose() + Entropy
for i in range(Entropy.shape[0]):
    Entropy[i, i] *= .5
```

```python
plt.matshow(Entropy, origin='lower', extent=[0, 2, 0, 2], cmap='viridis')
```

```python
C = np.loadtxt('Correlation.dat')
C = C.transpose() + C
for i in range(C.shape[0]):
    C[i, i] *= .5
plt.matshow(np.log(C), origin='lower', extent=[0, 2, 0, 2], cmap='viridis')
plt.colorbar()
plt.show()
```
