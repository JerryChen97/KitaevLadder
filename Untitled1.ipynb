```python
import numpy as np
```

```python
for Jx in np.linspace(2, 0, 201):
    N = int(np.round(Jx / 0.01))
    print(f'Jx = {Jx}, N=', N)
    Jy_list = np.linspace(Jx - 0.01, 0, N)
    for Jy in Jy_list:
        print(f'Jy = {Jy}')
```

```python
def to_index(Jx: float) -> int:
    return int(np.round(Jx/.01))

i = 201
j = 200
for J in np.linspace(2., 0, 201):
    j = to_index(J)
    assert i-j==1
    i = j
```
