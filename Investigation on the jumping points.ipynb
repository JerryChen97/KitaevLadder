In the issue https://github.com/JerryChen97/KitaevLadder/issues/32 we mentioned a usual phenomenon where the finite scaling does not happen normally all the time but suddenly jumps somewhere.

Now we need to first of all reproduce this phenomenon and then delve into the details.

```python
import numpy as np
import matplotlib.pyplot as plt

from scipy import optimize
```

```python
from kitaev_ladder_snake1 import run_atomic, save_after_run, finite_scaling, load_data, fDMRG_KL
```

```python
S = 1
Jx = 0.845
Jy = 0.845
Jz = 1.
```

```python
chi_list = [int(2**k) for k in np.linspace(3, 7, 11)]
# print(chi_list)
result = finite_scaling(S=S, Jx=Jx, Jy=Jy, Jz=Jz, chi_list=chi_list)
```

```python
result
```

```python
S_list = result[0]
xi_list = result[1]
logxi_list = np.log(xi_list)
plt.plot(logxi_list, S_list, 'o')
```

OK so we now can confirm that this always happens for the point `(0.845, 0.845, 1.)` for `S=1` case.

# Save them

```python
chi_list = [int(2**k) for k in np.linspace(3, 8, 16)]
print(chi_list)
```

```python
psi_list = []
energy_list = []

init_psi = None
for chi in chi_list:
    result = run_atomic(S=S, Jx=Jx, Jy=Jy, Jz=Jz, chi=chi, initial_psi=init_psi)
    
    psi = result['psi']
    psi_list.append(psi.copy())
    energy_list.append(result['energy'])
    init_psi = psi
    
```

```python
plt.plot(chi_list, energy_list, 'o-')
plt.xlabel('Bond dimension')
plt.ylabel('Energy')
plt.show()
```

```python
entropy_list = [np.mean(psi.entanglement_entropy()) for psi in psi_list]
plt.plot(chi_list, entropy_list, 'o-')
plt.xlabel('Bond dimension')
plt.ylabel('Entanglement Entropy')
plt.show()
```

```python
correlation_length_list = [psi.correlation_length() for psi in psi_list]
plt.plot(chi_list, correlation_length_list, 'o-')
plt.xlabel('Bond dimension')
plt.ylabel('Correlation Length')
plt.show()
```

```python
# plt.plot(np.log(correlation_length_list)[3:], )
```

```python
begin = 8
end = -1

effective_log_cl_list = np.log(correlation_length_list[begin:end])
effective_S_list = entropy_list[begin:end]
plt.plot(effective_log_cl_list, effective_S_list, 'o')
```

```python
def central_charge_func_logxi_to_S(logxi, c, a):
    return (c * logxi / 6) + a
```

```python
x_list = effective_log_cl_list
y_list = effective_S_list
parameters, covariances = optimize.curve_fit(central_charge_func_logxi_to_S, x_list, y_list)
c = parameters[0]
a = parameters[1]
```

```python
plt.clf()
# plot the scatters of raw data
plt.plot(x_list, y_list, 'o', label='iDMRG data')

# plot the curve fitting
xs = np.arange(x_list[0]-.1, x_list[-1]+.1, .01)
ys = [central_charge_func_logxi_to_S(x, c, a) for x in xs]
plt.plot(xs, ys, label=r'Curve fitting: $S=\frac{c}{6}\log\xi + a, c=%.2f, a=%.2f$' % (c, a))

plt.xlabel(r'$\log\xi$')
plt.ylabel(r'$S$')
plt.legend()
# plt.savefig("FiniteScaling"+naming(Jx, Jy, Jz, L))
plt.show()
```
