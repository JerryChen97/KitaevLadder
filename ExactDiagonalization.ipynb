I want a function to automatically calculate the degeneracy of each level.

```python
def calc_deg(list_sorted, err_tol=1e-7):
    """
        Calc the degeneracy dictionary of a sorted list.
        Input: a list which is sorted.
    """
    
    # initialize the algorithm states with the first element of the given sorted list
    previous = list_sorted[0]
    deg_dict = {}
    deg_dict[previous] = 1
    
    for i in range(1, len(list_sorted)):
        current = list_sorted[i]
        if current < previous:
            raise ValueError("The input list is not sorted ascendently!")
        if abs(current - previous) < err_tol: # then the previous one is still degenerate
            deg_dict[previous] += 1
        else: # then this is a new level
            deg_dict[current] = 1
            previous = current
            
    return deg_dict
```
