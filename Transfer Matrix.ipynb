When we are using iDMRG in MPS language, it is crucially important to determine whether the resulting wavefunction in MPS is reasonable or not.
One key property is whether or not a wavefunction is a cat state; an easiest example is that

$$|\psi\rangle \propto | \alpha \rangle + | \alpha \rangle$$
where $\alpha$ is a parameter used to describe certain property, e.g. a real number.

A classical example is the Transverse-Field Ising model (TFI):
$$ H_{TFI} = \sum_{\langle i j \rangle} \sigma^z_{i} \sigma^z_{j} + J \sum_i \sigma^x_i $$
where a well-known fact is that in the symmetry-breaking phase, i.e. $J < 1$, there will be 'two' degenerate states in the ground-state subspace of the whole Hilbert space.

In the extreme limit where $J=0$, the ground-state space can be easily written as 
$$\mathcal{H}_{GS} = \{ |\uparrow\uparrow\cdots\uparrow\rangle, |\downarrow\downarrow\cdots\downarrow\rangle \}$$
where $\sigma^z |\uparrow\rangle = |\uparrow\rangle, \sigma^z |\downarrow\rangle = -|\downarrow\rangle$ is the usual convention for a single spin.

The reality is that the ground states, viewed as a whole set, are not just two-fold degenerate, but all live in a subspace with dimension 2.
However, in this whole subspace, the two representatives `all-up` and `all-down` stand for the symmetry breaking essence of this phase, in the sense that
$$\Sigma^x|\uparrow\rangle = |\downarrow\rangle$$
$$\Sigma^x|\downarrow\rangle = |\uparrow\rangle$$
where $\Sigma^x = \prod_i \sigma^x_i$ is the global symmetry TFI possess (it commutes with the Hamiltonian).

Meanwhile, there is a local order parameter to distinguish between the two different states:
$$\langle \sigma^z \rangle = \pm 1$$

Now if we add the two states together, we will get a `cat-state`, for which the local order parameter can not work properly anymore; and in the same time, the correlation length of this state will be infinite.
In MPS language, this property can be expressed as the second eigenvalue of the transfer matrix is `1`.

## Generate all sets of the first two eigenvalues of transfer matrices

```python
eig1 = np.zeros((201, 201))
eig2 = np.zeros_like(eig1)
```

```python
for J, psi in wf_dict.items():
    x, y, z = J
    i, j = (int(np.round(x*100)), int(np.round(y*100))) # the coordinate
#     print(i, j)
    tm = TransferMatrix(psi, psi)
    tm_eigs = tm.eigenvectors(num_ev=2)[0]
    eig1_v = np.abs(tm_eigs[0])
    eig2_v = np.abs(tm_eigs[1])
    eig1[i, j] = eig1_v
    eig2[i, j] = eig2_v
    if i != j: # then symmetrize this value lol
        eig1[j, i] = eig1_v
        eig2[j, i] = eig2_v
```
